## Importando librerias

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
import seaborn as sb
```

```python
data = pd.read_csv("meteo.csv")
```

<p> Limpiando la basura del dataframe

```python
d = data.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm','WindDir9am','WindDir3pm'], 1)
```

```python
#Reemplazando el null por los promedios de cada paramtetro

d['MinTemp'].fillna(d['MinTemp'].mean(),inplace=True)
d['MaxTemp'].fillna(d['MaxTemp'].mean(),inplace=True)
d['Rainfall'].fillna(d['Rainfall'].mean(),inplace=True)
d['WindGustSpeed'].fillna(d['WindGustSpeed'].mean(),inplace=True)
d['WindSpeed9am'].fillna(d['WindSpeed9am'].mean(),inplace=True)
d['Humidity9am'].fillna(d['Humidity9am'].mean(),inplace=True)
d['Humidity3pm'].fillna(d['Humidity3pm'].mean(),inplace=True)
d['Pressure9am'].fillna(d['Pressure9am'].mean(),inplace=True)
d['Pressure3pm'].fillna(d['Pressure3pm'].mean(),inplace=True)
d['RISK_MM'].fillna(d['RISK_MM'].mean(),inplace=True)
d['Temp9am'].fillna(d['Temp9am'].mean(),inplace=True)
d['Temp3pm'].fillna(d['Temp3pm'].mean(),inplace=True)
```

```python
#Reemplazando el null por la moda de cada parametro
#No se realiza para ciertas variables debido a que NWW por ejemplo se repite en ellas y se mezclan

d['WindGustDir'].fillna(d['WindGustDir'].mode().get(0),inplace=True)
'''
d['WindDir9am'].fillna(d['WindDir9am'].mode().get(0),inplace=True)
d['WindDir3pm'].fillna(d['WindDir3pm'].mode().get(0),inplace=True)
d['WindSpeed3pm'].fillna(d['WindSpeed3pm'].mode().get(0),inplace=True)
'''
d['RainToday'].fillna(d['RainToday'].mode().get(0),inplace=True)
```

```python
# Generando las filas Dummies para parametros que son categorias
Location = pd.get_dummies(d['Location'],drop_first=True)
WindGustDir = pd.get_dummies(d['WindGustDir'],drop_first=True)
'''
WindDir9am = pd.get_dummies(d['WindDir9am'],drop_first=True)
WindDir3pm = pd.get_dummies(d['WindDir3pm'],drop_first=True)
'''
RainToday = pd.get_dummies(d['RainToday'],drop_first=True)
RainTomorrow = pd.get_dummies(d['RainTomorrow'],drop_first=True)

d = pd.concat([Location,WindGustDir,RainToday,RainTomorrow],axis=1)
```
